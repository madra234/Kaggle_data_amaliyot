```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot(sonlar)
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y, linewidth = 2, linestyle = '--', color = 'g', alpha = 0.7, marker = 'o')
plt.title('sin(x) funksiya')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.figure(figsize = (10, 5))
plt.show()
```

```python
df_plt = pd.read_csv('/content/04-VISUAL-AMALIYOT/wether.csv')
df_plt
df_plt.set_index('Year', inplace=True)
df_plt
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = df_plt['T']
plt.figure(figsize = (16, 8))
plt.plot(x, linewidth = 2, linestyle = '--', marker = 'h', markerfacecolor = 'r')
plt.title('Toshkent shahri uchun yillik o\'rtacha havo harorati ')
plt.xlabel('Yil')
plt.ylabel('Temperatura')

plt.show()
```

```python
maxT = df_plt['TM']
minT = df_plt['Tm']
plt.figure(figsize = (16, 8))
plt.plot(maxT, linewidth = 2, linestyle = '--', marker = 'h')
plt.plot(minT, linewidth = 2,  marker = 'v')
plt.legend(['O\'rtacha max harorat', 'O\'rtacha min harorat'], loc=4, fontsize = 'x-large')
plt.grid(True)
plt.title('Toshkent shahri uchun yillik o\'rtacha havo harorati ')
plt.xlabel('Yil')
plt.ylabel('Temperatura')
plt.savefig('tempMinMax.png')
plt.show()
plt.savefig('grafik.pdf')
```
